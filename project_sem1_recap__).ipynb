```python
#quadratic equation considering all cases
import math
def roots(a, b, c):
  disc= b*b - 4*a*c
  sqrt_val=math.sqrt(abs(disc))

  if (b*b > 4*a*c):
    print("real and distinct roots")
    print((-b + sqrt_val / 2*a), (-b - sqrt_val / 2*a))
  elif (b*b == 4*a*c):
    print('real and equal roots')
    print(-b/2*a)
  else:
    print("imaginary roots")
    print(-b/2*a, "+i", sqrt_val, "and", -b/2*a, "-i", sqrt_val)

a=int(input())
b=int(input())
c=int(input())

if (a==0):
  "inappropriate coefficients"
else:
  roots(a, b, c)
```

```python
#finds the number of unique words and counts the frequency of each word in a data set
def word_count(words):
    word_set = set(words)
    word_counts = {}
    for word in word_set:
        word_counts[word] = words.count(word)
    return word_counts
words = ['Red', 'Green', 'Red', 'Blue', 'Red', 'Red', 'Green']
print(word_count(words))
```

```python
#Sum of first 100 even numbers and first 100 odd numbers

i = 1
sumE = 0
sumO = 0
while i <=100:
  sumE=sumE+(i*2)
  sumO=sumO+(i*2-1)
  i=i+1
print("Sum of first 100 even numbers is",sumE,"\n"+"Sum of first 100 odd  numbers is",sumO)
```

```python
#Decimal to Binary

n = int(input("Enter Decimal number: "))
binX=0
posX=1
while n>0:
  digit = n%2
  binX=binX+digit*posX
  posX*=10
  n=n//2
print(f"Binary: {binX} ")
```

```python
#Strong Number

n = int(input("Enter a number: "))
temp = n
sum = 0
while temp>0:
  dig = temp%10
  f=1
  for i in range (1,dig+1):
    f=f*i
  sum=sum+f
  temp=temp//10
if sum ==n:
  print(n,"is a strong number")
else:
  print(n,"is not a strong number")
```

```python
#let's implement some sorting techniques
#bubble sort
from array import *
n=int(input("How many elements? "))
a=array('i', [])
for i in range(n):
  x=(int(input("Enter the element: ")))
  a.append(x)
print(a)
for i in range(len(a)-1):
  for j in range(len(a)-1):
  if a[j]>a[j+1]:
    temp=a[j]
    a[j]=a[j+1]
    a[j+1]=temp


#selection sort
from array import *
n=int(input("How many elements? "))
a=array('i', [])
for i in range(n):
  x=(int(input("Enter the element: ")))
  a.append(x)
print(a)
for i in range(len(a)):
  select=i
  for j in range(i+1, len(a)):
    if a[select]>a[j]:
      select=j
    a[select], a[i]=a[i], a[select]
print(a)
```
